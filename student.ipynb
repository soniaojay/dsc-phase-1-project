3.Identifying outliers in popularity, vote_average, and vote_count

```python
import numpy as np

# Calculating IQR for each column
Q1 = rotten_tomatoes_df[['popularity', 'vote_average', 'vote_count']].quantile(0.25)
Q3 = rotten_tomatoes_df[['popularity', 'vote_average', 'vote_count']].quantile(0.75)
IQR = Q3 - Q1

# Determining outliers based on IQR method
outliers_IQR = rotten_tomatoes_df[((rotten_tomatoes_df[['popularity', 'vote_average', 'vote_count']] < (Q1 - 1.5 * IQR)) | (rotten_tomatoes_df[['popularity', 'vote_average', 'vote_count']] > (Q3 + 1.5 * IQR))).any(axis=1)]

# Plotting mean lines and outliers
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(rotten_tomatoes_df['popularity'], bins=50, color='skyblue', edgecolor='black')
plt.axvline(rotten_tomatoes_df['popularity'].mean(), color='red', linestyle='dashed', linewidth=1)
plt.title('Popularity Distribution with Mean')
plt.scatter(outliers_IQR['popularity'], np.zeros_like(outliers_IQR['popularity']), color='red', marker='x')

plt.subplot(1, 3, 2)
plt.hist(rotten_tomatoes_df['vote_average'], bins=50, color='salmon', edgecolor='black')
plt.axvline(rotten_tomatoes_df['vote_average'].mean(), color='red', linestyle='dashed', linewidth=1)
plt.title('Vote Average Distribution with Mean')
plt.scatter(outliers_IQR['vote_average'], np.zeros_like(outliers_IQR['vote_average']), color='red', marker='x')

plt.subplot(1, 3, 3)
plt.hist(rotten_tomatoes_df['vote_count'], bins=50, color='green', edgecolor='black')
plt.axvline(rotten_tomatoes_df['vote_count'].mean(), color='red', linestyle='dashed', linewidth=1)
plt.title('Vote Count Distribution with Mean')
plt.scatter(outliers_IQR['vote_count'], np.zeros_like(outliers_IQR['vote_count']), color='red', marker='x')

plt.show()
```

Identify Top Performing Films based on the composite score which is weight between Popularity, vote_count and vote_average.

```python
# Normalize the data
cleaned_rotten_tomatoes_df.loc[:, 'normalized_popularity'] = (cleaned_rotten_tomatoes_df['popularity'] - cleaned_rotten_tomatoes_df['popularity'].min()) / (cleaned_rotten_tomatoes_df['popularity'].max() - cleaned_rotten_tomatoes_df['popularity'].min())
cleaned_rotten_tomatoes_df.loc[:, 'normalized_vote_count'] = (cleaned_rotten_tomatoes_df['vote_count'] - cleaned_rotten_tomatoes_df['vote_count'].min()) / (cleaned_rotten_tomatoes_df['vote_count'].max() - cleaned_rotten_tomatoes_df['vote_count'].min())
cleaned_rotten_tomatoes_df.loc[:, 'normalized_average_vote'] = (cleaned_rotten_tomatoes_df['vote_average'] - cleaned_rotten_tomatoes_df['vote_average'].min()) / (cleaned_rotten_tomatoes_df['vote_average'].max() - cleaned_rotten_tomatoes_df['vote_average'].min())

# Assign weights to each metric
weight_popularity = 0.5
weight_vote_count = 0.3
weight_average_vote = 0.2

# Calculate the composite score
cleaned_rotten_tomatoes_df.loc[:, 'composite_score'] = (weight_popularity * cleaned_rotten_tomatoes_df['normalized_popularity']) + (weight_vote_count * cleaned_rotten_tomatoes_df['normalized_vote_count']) + (weight_average_vote * cleaned_rotten_tomatoes_df['normalized_average_vote'])

# Rank the films based on the composite score
top_movies = cleaned_rotten_tomatoes_df.sort_values(by='composite_score', ascending=False)

# Select top performing films (e.g., top 10)
top_performing_movies = top_movies.head(10)

# Display the top performing films
print("Top 10 Performing Movies based on Composite Score:")
print(top_performing_movies[['title', 'popularity', 'vote_count', 'vote_average', 'composite_score','release_month','release_year']])
```
